Using Python, create a class named pet with attributes name and gender. Using the 
constructor, set the values of these attributes. Define a method named display() that 
prints the name and gender of the pet. Create two objects for the pet class; one object 
for a dog and one for a cat. Provide appropriate values to the attributes. 

```python
class Pet:
    def __init__(self, name, gender):
        self.name = name 
        self.gender = gender 
    def display(self):
        print(f"The name of the pet is {self.name} and the gender is {self.gender}")
        

pet1 = Pet("Dengo", "Male")
pet2 = Pet("Zumi", "Male")

pet1.display()
pet2.display()
```

Create a class named Order with attributes product_mount and price. Using 
the constructor, set the values of these attributes. Define a method named calculate_bill() 
that returns the total bill for the Order (product_amount*price). Create two objects for 
the order class and provide appropriate values to the attributes. Print the total bill for 
both objects.

```python
class Order:
    def __init__(self, product_mount, price): 
        self.product_mount = product_mount
        self.price = price
    def calculate_bill(self):
        total_bill = self.product_mount*self.price
        return total_bill
    
order1 = Order(3, 15.90)
order2 = Order(20, 3.50)


x = order1.calculate_bill()
y = order2.calculate_bill()

print(f"The total bill of order 1 is {x} and of order 2 is {y}")
```
